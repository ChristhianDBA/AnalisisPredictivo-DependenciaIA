```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt

# Suponiendo que ya tienes el dataframe 'nuevodf1'

# Mapeo de los años de estudio
mapa_ano = {
    '1er año de estudios universitarios': 1,
    '2do año de estudios universitarios': 2,
    '3er año de estudios universitarios': 3,
    '4to año de estudios universitarios': 4,
    '5to año de estudios universitarios': 5
}



# Asegurarse de que las columnas de utilidad sean numéricas
columnas_utilidades_ia = ['Fuente', 'Formulacion', 'Prueba', 'Diseño', 'Ahorro', 'Aclarar', 'Resumen', 'Traduccion',
                          'Citas', 'Codificacion', 'Revision', 'Graficos', 'Presentaciones']

# Agrupar los datos por 'Año' (aquí ya no necesitas convertir a fecha)
utilidad_chatgpt_agrupada = nuevodf1.groupby('Año')['Utilidad_ChatGPT'].mean()
utilidades_ia_agrupadas = nuevodf1.groupby('Año')[columnas_utilidades_ia].mean()

# Función para aplicar Holt-Winters y proyectar en el futuro
def aplicar_holt_winters(datos, periodos_futuros=5):
    modelo_hw = ExponentialSmoothing(datos, trend="add", seasonal=None, damped_trend=False)
    modelo_ajustado = modelo_hw.fit()
    proyeccion = modelo_ajustado.forecast(periodos_futuros)
    return proyeccion

# Proyección para 'Utilidad_ChatGPT'
proyeccion_utilidad_chatgpt = aplicar_holt_winters(utilidad_chatgpt_agrupada)

# Proyección para las utilidades específicas de IA (cada columna de la lista)
proyecciones_utilidades_ia = {}
for columna in columnas_utilidades_ia:
    proyecciones_utilidades_ia[columna] = aplicar_holt_winters(utilidades_ia_agrupadas[columna])

# Visualización de los resultados
plt.figure(figsize=(14, 8))

# Plot para 'Utilidad_ChatGPT'
plt.subplot(2, 1, 1)
plt.plot(utilidad_chatgpt_agrupada, label="Datos Históricos - Utilidad ChatGPT")
plt.plot(range(utilidad_chatgpt_agrupada.index[-1] + 1, utilidad_chatgpt_agrupada.index[-1] + 1 + 5),
         proyeccion_utilidad_chatgpt, color="red", linestyle="--", label="Proyección - Utilidad ChatGPT")
plt.xlabel("Año de Estudio")
plt.ylabel("Calificación Promedio de Utilidad ChatGPT")
plt.title("Proyección de la Utilidad de ChatGPT en el Tiempo")
# Mover la leyenda a la derecha
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot para utilidades específicas de IA
plt.subplot(2, 1, 2)
for columna, proyeccion in proyecciones_utilidades_ia.items():
    plt.plot(utilidades_ia_agrupadas[columna], label=f"Datos Históricos - {columna}")
    plt.plot(range(utilidades_ia_agrupadas.index[-1] + 1, utilidades_ia_agrupadas.index[-1] + 1 + 5),
             proyeccion, linestyle="--", label=f"Proyección - {columna}")

plt.xlabel("Año de Estudio")
plt.ylabel("Calificación Promedio de Utilidades Específicas")
plt.title("Proyección de las Utilidades Específicas de IA en el Tiempo")
# Mover la leyenda a la derecha
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar el diseño para que no se sobreponga la leyenda
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix



# Definir las variables independientes (X) y dependiente (y)
X = nuevodf1[['Utilidad_ChatGPT', 'Trampa_ChatGPT', 'Ahorro', 'Aclarar', 'Codificacion',
              'Resumen', 'Traduccion', 'Citas', 'Graficos', 'Presentaciones']]
y = nuevodf1['Usa_IA']  # 0 = No, 1 = Sí (se asume que esta columna tiene valores binarios)

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Realizar predicciones
y_pred = rf_model.predict(X_test)

# Evaluar el modelo
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Obtener importancia de las características
importances = rf_model.feature_importances_
features = X.columns

# Graficar las importancias
plt.figure(figsize=(10, 6))
plt.barh(features, importances, color='skyblue')
plt.xlabel("Importancia de las Características")
plt.ylabel("Características")
plt.title("Importancia de las Características según Random Forest")
plt.show()
```

```python
# Importar librerías necesarias
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt


# Definir variables independientes (features) y la variable objetivo (target)
X = nuevodf1.drop('Usa_IA', axis=1)  # Reemplaza 'Usa_IA' con el nombre de tu columna objetivo
y = nuevodf1['Usa_IA']

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Hacer predicciones
y_pred = rf_model.predict(X_test)

# Evaluar el modelo
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Graficar matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.title('Matriz de Confusión')
plt.show()

# Graficar importancia de las características
importances = rf_model.feature_importances_
features = X.columns
plt.figure(figsize=(10, 6))
plt.barh(features, importances, color='skyblue')
plt.xlabel("Importancia de las Características")
plt.ylabel("Características")
plt.title("Importancia de las Características según Random Forest")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix



# Selección de variables (independientes y dependiente)
# Variable dependiente: Passed_exams
y = nuevodf2['Examenes_aprobados']

# Variables independientes (puedes ajustar esta lista)
X = nuevodf2[['Conocimiento de IA', 'Valoración de IA_Educación',
              'Ventaja de IA_Enseñanza', 'Ventaja de IA_Aprendizaje',
              'Desventaja de IA_Proceso Educativo', 'Promedio_de_notas']]

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy:.2f}')

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión')
plt.show()

# Visualización del impacto de cada variable independiente
coef_df = pd.DataFrame({'Variable': X.columns, 'Coeficiente': model.coef_[0]})
coef_df = coef_df.sort_values(by='Coeficiente', ascending=False)
sns.barplot(data=coef_df, x='Coeficiente', y='Variable')
plt.title('Importancia de las variables')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Selección de variables
X = nuevodf2[['Promedio_de_notas']].values  # Variable independiente
y = nuevodf2['Examenes_aprobados'].values  # Variable dependiente

# Crear el modelo de regresión logística
model = LogisticRegression()
model.fit(X, y)

# Predicciones para la curva
X_range = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
y_pred_prob = model.predict_proba(X_range)[:, 1]  # Probabilidad de aprobar

# Gráfico
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales', alpha=0.7)
plt.plot(X_range, y_pred_prob, color='red', linewidth=2, label='Curva logística')
plt.title('Regresión Logística: Promedio de Notas vs Exámenes Aprobados')
plt.xlabel('Promedio de Notas')
plt.ylabel('Probabilidad de Aprobar (Exámenes)')
plt.legend()
plt.grid(True)
plt.show()
```
